```python
global_temp_country = pd.read_csv('./data/GlobalLandTemperaturesByCountry.csv')
print(global_temp_country.__len__())

france_temperatures = global_temp_country.loc[(global_temp_country["Country"] == "France") & (global_temp_country["dt"] > "2000-01-01") ]
# global_temp_country = global_temp_country.drop(['dt', "Country"], axis=1)
# france_temperatures.plot(x='AverageTemperature',y='AverageTemperatureUncertainty',kind='scatter')
france_temperatures.plot(x='dt', marker='o', markerfacecolor='red')
```
